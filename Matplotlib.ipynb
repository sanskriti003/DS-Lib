# Using ggplot

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,2,3]
y=[23,44,32]
x2=[26,66,43]
y2=[33,45,77]
plt.plot(x,y,label='first',linewidth=3)#to increase the width of line
plt.plot(x2,y2,label='second')
plt.legend()#shows top right first and second thing of colour of lines
plt.show()
```

to use scatterplot

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,2,3]
y=[23,44,32]
x2=[26,66,43]
y2=[33,45,77]
plt.scatter(x,y,label='first',linewidth=3)#to increase the width of line
plt.scatter(x2,y2,label='second')
plt.legend()#shows top right first and second thing of colour of lines
plt.show()
```

to use histogram

```python
from matplotlib import pyplot as plt
from matplotlib import style
style.use('ggplot')
x=[1,2,3]
y=[23,44,32]
x2=[26,66,43]
y2=[33,45,77]
plt.hist(x,label='first',linewidth=3)#to increase the width of line
plt.hist(x2,label='second')
plt.legend()#shows top right first and second thing of colour of lines
plt.show()
```
